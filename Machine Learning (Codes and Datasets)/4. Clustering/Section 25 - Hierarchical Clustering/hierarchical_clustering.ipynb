# Hierarchical Clustering

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('Mall_Customers.csv')
X = dataset.iloc[:, [3, 4]].values
```

## Using the dendrogram to find the optimal number of clusters

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X, 'ward'))

plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidiean Distance')
plt.show()
```

## Training the Hierarchical Clustering model on the dataset

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters=5,linkage='ward',affinity='euclidean')
y_hc = hc.fit_predict(X)
```

```python
print(y_hc)
```

## Visualising the clusters

```python
# Visualising the clusters
plt.figure(figsize=(10, 6))
plt.scatter(X[y_hc == 0, 0],X[y_hc == 0,1],s = 100,label = 'Cluster 1', color='red')
plt.scatter(X[y_hc == 1, 0],X[y_hc == 1,1],s = 100,label = 'Cluster 2', color='blue')
plt.scatter(X[y_hc == 2, 0],X[y_hc == 2,1],s = 100,label = 'Cluster 3', color='green')
plt.scatter(X[y_hc == 3, 0],X[y_hc == 3,1],s = 100,label = 'Cluster 4', color='yellow')
plt.scatter(X[y_hc == 4, 0],X[y_hc == 4,1],s = 100,label = 'Cluster 5', color='magenta')
plt.title("Cluster of Customers")
plt.xlabel('Annual income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.gca().spines['right'].set_color('none')
plt.gca().spines['top'].set_color('none')
plt.show()
```
